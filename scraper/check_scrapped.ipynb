```python
import pandas as pd
from pathlib import Path
import os
df = pd.read_pickle(r"data/stocktwits_raw_full.pkl")
df
```

```python
import re
# given a text, find the related symbols and remove it 
# return: a string of text and a list of symbols
def extract_syb(text):
    symbols = re.findall(r"\B[$][a-zA-Z]*", text)
    symbols = [i for i in symbols if i != "$"]
    if len(symbols) == 1:
        return symbols[0]
    else:
        return None

df['symbols'] = df['text'].apply(extract_syb)
df = df.dropna(subset=['symbols'])
df['symbols'] = [i.upper() for i in df['symbols']]
df
```

```python
not_yet_scrapped = set(stocks.ticker) - set([i[1:] for i in df.symbols])  
for i in not_yet_scrapped:
    print(i, end=' ')
```
