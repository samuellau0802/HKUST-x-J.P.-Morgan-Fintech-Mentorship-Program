```python
from transformers import pipeline

# Allocate a pipeline for sentiment-analysis
classifier = pipeline('sentiment-analysis')
```

```python
import pandas as pd
from textblob import TextBlob
df = pd.read_csv("data/stockerbot-export.csv", error_bad_lines=False)
df.head()
```

```python
sample = df.sample(20)
for i in sample['text']:
    print(i)
    print(TextBlob(i).sentiment)
    print(classifier(i))
    print()
```
